```python
import matplotlib.pyplot as plt
import seaborn as sns
 # Wider figure

# Filter wildtype and mutant
wt_data = data_g[data_g['mut'] == 'wt']['days_to_flowering']
mut_data = data_g[data_g['mut'] != 'wt']['days_to_flowering']

# Compute WT mean and std
wt_mean = wt_data.mean()
wt_std = wt_data.std()

# Compute ±3 std lines
std_lines = [wt_mean + i * wt_std for i in range(-3, 4)]

# Find min in mutant group
mut_min = mut_data.min()
mut_min_std = (mut_min - wt_mean) / wt_std

# Set plot style
sns.set(style="white")

# Create figure
plt.figure(figsize=(16, 6)) 

# Histogram + KDE for both groups
sns.histplot(data=data_g, x='days_to_flowering', hue='mut', bins=70, 
             kde=True, palette={'wt': 'grey', 'mut': 'steelblue'}, edgecolor=None)

# Offset for label positioning
label_offset = wt_std * 0.05

# Add std lines
for i, val in zip(range(-3, 4), std_lines):
    plt.axvline(val, color='grey', linestyle='--', linewidth=1)
    plt.text(val + label_offset, plt.ylim()[1] * 0.9, f'{i} std',
             rotation=90, ha='left', va='top', color='grey')

# Mark the min mutant flowering time
plt.axvline(mut_min, color='red', linestyle='-', linewidth=2 )
plt.text(mut_min + label_offset, plt.ylim()[1] * 0.95,
         f'Mut min\n({mut_min_std:.2f}σ)', rotation=90, ha='left', va='top', color='red')

sns.despine(top=True, right=True)



plt.xlabel("Days to flowering")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig('flowering_time_FLC_nockouts.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")

# Violin plot (vertical by default when x is numeric)
sns.violinplot(data=data_g, x='mut', y='days_to_flowering', inner=None)

# Add individual data points
sns.stripplot(data=data_g, x='mut', y='days_to_flowering', 
              color='black', size=3, jitter=True, alpha=0.6)

# Clean up
plt.xlabel("Genotype")
plt.ylabel("Days to Flowering")
sns.despine()
plt.tight_layout()
plt.show()
```
