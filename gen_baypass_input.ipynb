```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import math 
import os

import allel
import subprocess
```

```python
allele_counts_slim_full_cov = pd.read_csv('allele_counts_slim_full_cov.csv').set_index('Unnamed: 0')
```

```python
allele_counts_slim_full_cov
```

```python
subps = allele_counts_slim_full_cov.columns
```

```python
subps = [i for i in subps if 'ref_' in i]
```

```python
pool_size = []
for subp in subps:
    leng = len(allele_counts_slim_full_cov[subp].dropna())
    pool_size.append(leng)
```

```python
#.str.replace('ref_', '')
subps = [i.replace('ref_', '') for i in subps]
```

```python
pool_size = pd.Series(pool_size, subps)
```

```python
pool_size.loc['subp0']
```

```python
allele_counts_slim_full_cov.columns
```

```python
for col in allele_counts_slim_full_cov.columns:
    if 'ref_' in col: 
        allele_counts_slim_full_cov.loc[:,col] = allele_counts_slim_full_cov.loc[:, col].fillna(pool_size.loc[col[4:]])
    else:
        allele_counts_slim_full_cov.loc[:,col] = allele_counts_slim_full_cov.loc[:, col].fillna(0)
```

```python
allele_counts_slim_full_cov
```

```python
allele_counts_slim_full_cov = allele_counts_slim_full_cov.astype(int)
```

```python
allele_counts_slim_full_cov.to_csv('baypass/geno_table_slim', header=None, sep = ' ', index=False)
```

```python
# double the pool size because it is asking for number of chromosomes 
pool_size = pool_size * 2
```
